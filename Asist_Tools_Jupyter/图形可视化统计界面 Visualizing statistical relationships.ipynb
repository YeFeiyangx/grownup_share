```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.animation import FuncAnimation  # 动图的核心函数
```

```python
## relplot()
## This is a figure-level function for visualizing statistical relationships using two common approaches: scatter plots and line plots.
## 用散点和线性图像对统计信息做可视化描述
## relplot() combines a FacetGrid with one of two axes-level functions
## 这是一个二维表格描述
```

```python
## hue : name in data, optional
## Grouping variable that will produce elements with different colors. 
## Can be either categorical or numeric, although color mapping will behave differently in latter case.
## 用于第三维度数据，可以用不同颜色表示不同数据类

## style : name in data, optional
## Grouping variable that will produce elements with different styles. Can have a numeric dtype but will always be treated as categorical.
## 用于第三维度数据，可以用不同展现方式表示数据点

```

```python
tips = sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
```

```python
## hue 表示smoker项中，是否吸烟的人用不同颜色来表示。
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
## style 表示smoker项中，是否吸烟的人用不同颜色表示，吸烟和不吸烟人的各自人群中用时间来区分。
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time",data=tips);
```

```python
## style 表示smoker项中，是否吸烟的人用不同颜色表示。
## 吸烟和不吸烟人的各自人群中用时间来区分。
## 图形形状可进一步引入图标尺寸作为新的数据维度。
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time",size="size", data=tips);
```

```python
## style 表示smoker项中，是否吸烟的人用不同颜色表示。
## 吸烟和不吸烟人的各自人群中用时间来区分。
## 图形形状可进一步引入图标尺寸作为新的数据维度。
## sizes可以控制用图形大小表示新的数据维度的范围。
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time",size="size",sizes=(15,200), data=tips);
```

```python
## Emphasizing continuity with line plots
## 用线性图来表示连续性
```

```python
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))# 用两字典转为数组，key为列名，value为数据
print(df.shape,"\n",df.head())
# g = sns.relplot(x="time", y="value", kind="line", data=df)
# g.fig.autofmt_xdate()
sns.relplot(x="time", y="value", kind="line", data=df)
```

```python
df = pd.DataFrame(np.random.randn(5, 2).cumsum(axis=0), columns=["x", "y"])# 不对随机生成的数组进行排序，直接用原每行按列累加的坐标排序。
sns.relplot(x="x", y="y", sort=False, kind="line", data=df);
```

```python
## Plotting subsets of data with semantic mappings
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event",
            dashes=False, markers=True, kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
            kind="line", data=fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="region",
            units="subject", estimator=None,
            kind="line", data=fmri.query("event == 'stim'"));
```

```python
dots = sns.load_dataset("dots").query("align == 'dots'")
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            kind="line", data=dots);
```

```python
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            palette=palette,
            kind="line", data=dots);
```

```python
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(x="time", y="firing_rate",
            hue="coherence", style="choice",
            hue_norm=LogNorm(),
            kind="line", data=dots);
```

```python
sns.relplot(x="time", y="firing_rate",
            size="coherence", style="choice",
            kind="line", data=dots);
```

```python
sns.relplot(x="time", y="firing_rate",
           hue="coherence", size="choice",
           palette=palette,
           kind="line", data=dots);
```
